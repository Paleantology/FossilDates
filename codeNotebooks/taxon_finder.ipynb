```python
import pandas as pd
import dendropy
import os
import sys
```

```python
barden_tree = dendropy.Tree.get_from_path("../Data/accessory/666TipTree.tre", schema = "newick", preserve_underscores = True)
estimated_tree = dendropy.Tree.get_from_path("../Data/accessory/init.tre", schema = "newick", preserve_underscores = True)
```

```python
for t in barden_tree.taxon_namespace:
    print(t.label)
```

```python
est_label = []

for tax in barden_tree.taxon_namespace:
    est_label.append(tax.label)
```

```python
shared_labs = []

for lab in estimated_tree.taxon_namespace:
    if lab.label in est_label:
        shared_labs.append(lab.label)
```

```python
barden_tree.retain_taxa(shared_labs)
```
